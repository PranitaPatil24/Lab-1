```python
!pip install beautifulsoup4 
import os
from bs4 import BeautifulSoup

folder = r"C:\Lab_1\ISR\website"
index = {}

# Build index
for f in os.listdir(folder):
    if f.endswith(".html"):
        text = BeautifulSoup(open(os.path.join(folder, f), encoding="utf-8").read(), "html.parser").get_text().lower()
        for w in text.split():
            index.setdefault(w, set()).add(f)

# Search loop
while True:
    word = input("Search word (or 'exit'): ").lower()
    if word == "exit":
        break
    print("Found in:", ", ".join(index.get(word, [])) or "No page found")
```

```python
#2nd code   --optional 
from bs4 import BeautifulSoup
import os

folder = r"C:\Lab_1\ISR\website"
pages = {}

for filename in os.listdir(folder):
    if filename.endswith(".html"):
        path = os.path.join(folder, filename)
        with open(path, 'r', encoding='utf-8') as f:
            soup = BeautifulSoup(f, "html.parser")
            text = soup.get_text(separator=" ", strip=True)
            pages[filename] = text.lower() 
```

```python
import re
index = {}

for page, content in pages.items():
    words = re.findall(r'\w+', content)
    for word in words:
        if word not in index:
            index[word] = set()
        index[word].add(page)
```

```python
query = input("Enter search query: ").lower()
words = query.split()
results = set()

for word in words:
    if word in index:
        if not results:
            results = set(index[word])
        else:
            results = results.intersection(index[word])

# Check final results after processing all words
if results:
    print("Pages containing your query:", results)
else:
    print("No pages found.")
```
